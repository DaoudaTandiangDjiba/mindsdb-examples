```python
import mindsdb

# Instantiate a mindsdb Predictor
mdb = mindsdb.Predictor(name='btc-price')

# We tell the Predictor what column or key we want to learn and from what data
mdb.learn(
    from_data="dataset/bitcoin_train.csv", # the path to the file where we can learn from, (note: can be url)
    to_predict='price(USD)', # the column we want to learn to predict given all the data in the file
)
```

```python
from mindsdb import Predictor

# use the model to make predictions
result = Predictor(name='btc-price').predict(when={'txVolume(USD)': 6739584540.73, 'adjustedTxVolume(USD)': 3868097401.91,
    'txCount': 204913, 'exchangeVolume(USD)': 7394019840, 'generatedCoins': 1875, 'fees': 35.900, 'blockCount':  150})
print(result[0])
```
